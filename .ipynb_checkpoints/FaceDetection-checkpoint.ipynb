## Import openCV and assign HaarCascade

```python
import cv2
faceCascade = cv2.CascadeClassifier('haarcascade_frontalcatface_extended.xml')
```

```python
#Read Image 
image = cv2.imread('./test images/the man.jpeg')
```

```python
from matplotlib import pyplot as plt
plt.imshow(image)
```

```python
#Convert BGR image to RGB
image_RGB = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_RGB)
```

```python
#Convert to GrayScale
image_GR = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_GR)
```

```python
faces = faceCascade.detectMultiScale(
    image_GR,
    scaleFactor = 1.2,
    minNeighbors = 5,
    minSize = (30,30)
)

print(faces)
```
